Bivarite analysis

```python
sns.lineplot(data=df,x="carat",y="price")
plt.title("Carat vs Price")
plt.show()
```

```python
sns.regplot(data=df, x="carat", y="price")
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x="carat", y="price")
sns.regplot(data=df, x="carat", y="price", scatter=False, color="red")
plt.title("Carat vs Price")
plt.show()
```

```python
sns.barplot(data=df,x="clarity",y="carat")
plt.title("Carat vs clarity")
plt.show()
```

```python
sns.barplot(data=df,x="cut",y="price")
plt.title("Cut vs price")
plt.show()
```

```python
sns.barplot(data=df,x="cut",y="depth")
plt.title("Cut vs depth")
plt.show()
```

```python
sns.barplot(data=df,x="clarity",y="cut")
plt.title("Cut vs clarity")
plt.show()
```

```python
sns.countplot(data=df, x="clarity", hue="cut")
```
