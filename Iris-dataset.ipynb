```python
# PLotting Dataset on plots

#Line plots
fig, axes = plt.subplots(1, 2, figsize=(12, 5))  # 1 row, 2 columns

# --- Plot 1 ---
sns.lineplot(ax=axes[0], x="sepal_length", y="petal_width", data=df)
axes[0].set_title("(sepal_length) vs (petal_width)")

# --- Plot 2 ---
sns.lineplot(ax=axes[1], x="petal_length", y="sepal_width", data=df)
axes[1].set_title("(petal_length) vs (sepal_width)")

plt.tight_layout()
plt.show()
```

```python
x=df.columns
print(x)

fig, axes = plt.subplots(2, 2, figsize=(15, 8))

# 1
axes[0, 0].hist(df["sepal_length"], bins=10, edgecolor="black", linewidth=2)
axes[0, 0].set_title("Sepal Length")
axes[0, 0].set_xlabel("Values")
axes[0, 0].set_ylabel("Frequency")

# 2
axes[0, 1].hist(df["sepal_width"], bins=10, edgecolor="black", linewidth=2)
axes[0, 1].set_title("Sepal Width")
axes[0, 1].set_xlabel("Values")
axes[0, 1].set_ylabel("Frequency")

# 3
axes[1, 0].hist(df["petal_length"], bins=10, edgecolor="black", linewidth=2)
axes[1, 0].set_title("Petal Length")
axes[1, 0].set_xlabel("Values")
axes[1, 0].set_ylabel("Frequency")

# 4
axes[1, 1].hist(df["petal_width"], bins=10, edgecolor="black", linewidth=2)
axes[1, 1].set_title("Petal Width")
axes[1, 1].set_xlabel("Values")
axes[1, 1].set_ylabel("Frequency")

plt.tight_layout()
plt.suptitle("Histogram")
plt.show()
```

```python
#Pairplot

sns.pairplot(df,hue="species",kind="hist")
```

```python
#Boxplot

sns.boxplot(df)
```

```python
#Violin PLot

sns.violinplot(df,x="species",y='sepal_length')
```

```python
#Heatmap
data=df.drop(columns = ['species']).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(data,cmap="tab20c",annot=True)
```

```python
# Regression Line Plot

sns.lmplot(df,x='sepal_length', y='sepal_width',hue='species')
```
